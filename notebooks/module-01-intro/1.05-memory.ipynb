```python
from dotenv import load_dotenv
from pprint import pprint

load_dotenv()
```

## No memory

```python
from langchain.agents import create_agent


agent = create_agent(
    "gpt-5.4-nano"
)
```

```python
from langchain.messages import HumanMessage

question = HumanMessage(content="Hello my name is Tom and my favorite food is Pizza!")

response = agent.invoke(
    {"messages": [question]} 
)
```

```python
pprint(response['messages'][-1].content)
```

```python
question = HumanMessage(content="What's my favourite food?")

response = agent.invoke(
    {"messages": [question]} 
)
```

```python
pprint(response['messages'][-1].content)
```

## Memory

```python
from langgraph.checkpoint.memory import InMemorySaver  


agent = create_agent(
    "gpt-5.4-nano",
    checkpointer=InMemorySaver(),  
)
```

```python
from langchain.messages import HumanMessage

question = HumanMessage(content="Hello my name is Tom and my favorite food is Pizza!")
config = {"configurable": {"thread_id": "1"}}

response = agent.invoke(
    {"messages": [question]},
    config,  
)
```

```python
pprint(response['messages'][-1].content)
```

```python
question = HumanMessage(content="What's my favourite food?")

response = agent.invoke(
    {"messages": [question]},
    config,  
)
```

```python
pprint(response['messages'][-1].content)
```
