```python
from dotenv import load_dotenv

load_dotenv()
```

```python
from langchain.chat_models import init_chat_model
```

## Customizing your Model

```python
model = init_chat_model(
    model="gpt-5.4-nano",
    # Kwargs passed to the model:
    temperature=.5,
    # max_tokens=600,
    # timeout=.1,
    # max_retries=3,
)

response = model.invoke("How similar are Portuguese and Spanish?")
print(response.content)
```

## Model Providers


https://docs.langchain.com/oss/python/integrations/chat


```python
model = init_chat_model(model="claude-sonnet-4-5")

response = model.invoke("How similar are Portuguese and Spanish?")
print(response.content)
```

## Initialising and invoking an agent


```python
from langchain.agents import create_agent

agent = create_agent(model=model)
```

```python
agent = create_agent(model="claude-sonnet-4-5")
```

```python
agent = create_agent("gpt-5.4-nano")
```

```python
from langchain.messages import HumanMessage

response = agent.invoke(
    {"messages": [HumanMessage(content="How similar are Portuguese and Spanish?")]}
)
```

```python
from pprint import pprint

pprint(response)
```

```python
print(response['messages'][-1].content)
```

```python
from langchain.messages import AIMessage

response = agent.invoke(
    {"messages": [HumanMessage(content="How similar are Portuguese and Spanish?"),
    AIMessage(content="They are not similar at all."),
    HumanMessage(content="Interesting, tell me more about it.")]}
)

pprint(response)
```

```python
print(response['messages'][-1].content)
```

## Streaming Output


```python
for token, metadata in agent.stream(
    {"messages": [HumanMessage(content="Tell me all about the differences between Portuguese and Spanish"),]},
    stream_mode="messages"
):

    # token is a message chunk with token content
    # metadata contains which node produced the token

    if token.content:  # Check if there's actual content
        print(token.content, end="", flush=True)  # Print token
```
